## Importing neccessary libraries:

```python
import pandas as pd # for working with data!
import matplotlib.pyplot as plt # for visualizing the data!
import numpy as np # for calculating the moving average!
```

## Importing the Data:

```python
global_temp = pd.read_csv('global_data.csv') # importing 'global tempreature data'
ranchi_temp = pd.read_csv('ranchi_data.csv') # importing '~Ranchi tempreature data' : data for Ranchi over multiple year
patna_temp = pd.read_csv('patna_data.csv') # importing '~Patna tempreature data' : data for Patna over multiple year
bangalore_temp = pd.read_csv('bangalore_data.csv') # importing '~Bangalore tempreature data' : data for Banglore over multiple year
```

## 'Moving Average' for the Global data and Local data ('Ranchi','Patna'&'Bangalore') 

Moving averages are used to smoothen out data and to make it easier to observe long term trends and to ignore daily fluctuations.

```python
global_mv_avg = global_temp['avg_temp'].rolling(10).mean()
ranchi_mv_avg = ranchi_temp['avg_temp'].rolling(10).mean()
patna_mv_avg = patna_temp['avg_temp'].rolling(10).mean()
bangalore_mv_avg = bangalore_temp['avg_temp'].rolling(10).mean()
```

## Visualization using matplotlib library:

Data:
1. Global Average Temprature
2. Ranchi Average Temprature
3. Patna Average Temprature
4. Bangalore Average Temprature

```python
#Global Average Temprature
plt.plot(global_temp['year'],global_mv_avg,label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("Global Average Temperature")
plt.show()
```

```python
#Ranchi Average Temprature
plt.plot(ranchi_temp['year'],ranchi_mv_avg,label='Ranchi')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("Ranchi City Average Temperature")
plt.show()
```

```python
#Patna Average Temprature
plt.plot(patna_temp['year'],patna_mv_avg,label='Patna')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("Patna City Average Temperature")
plt.show()
```

```python
#Bangalore Average Temprature
plt.plot(bangalore_temp['year'],bangalore_mv_avg,label='Bangalore')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("Bangalore City Average Temperature")
plt.show()
```
